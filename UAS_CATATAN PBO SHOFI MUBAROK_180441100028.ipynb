# CLASS VS OBJECT

__A.PENGERTIAN CLASS__

    Class adalah struktur data yang bisa kita gunakan untuk mendefinisikan objek yang menyimpan data bersama nilai-nilai dan perilaku (behavior). Class juga merupakan suatu entitas yang merupakan bentuk program dari suatu abstraksi untuk permasalahan dunia nyata dan instans dari class merupakan realisasi dari beberapa objek. Class digunakan hanya untuk membuat kerangka dasar. Sebagai analogi, class bisa diibaratkan dengan laptop, mobil, orang dan lain sebagainya.
    
__B.BAGIAN DARI CLASS__
1.	Property

	Property (biasa disebut dengan atribut ) adalah data yang terdapat dalam sebuah class. Dalam contoh kali ini adalah property dari laptop bisa berupa merk, warna, jenis processor, ukuran layar, dll.
2.	Method

	Method adalah tindakan yang bisa dilakukan di dalam class. Dalam hal ini method dari class laptop adalah menghidupkan laptop, mematikan laptop, dan berbagai tindakan yang lain.
3.	Object

	Object adalah hasil cetak dari suatu class. Object berfungsi untuk memanggil semua property dan method yang ada dalam suatu class untuk di printkan.

Seperti contoh diatas :
	- Class	= mhs (Mahasiswa)
	- Method 	= tampilkan (menampilkan nama dan NIM)
		   tampilkanipk (menampilkan IPK)
	- Property	= Nama, NIM, IPS1, IPS2, IPS3
	- Object 	= Objek (berisikan Nama, NIM, IPS1, IPS2, IPS3)


# INHERITANCE

__A.PENGERTIAN INHERITANCE__

Inheritace adalah sebuah konsep yang penting di dalam python. Inheritance adalah sebuah proses dimana sebuah class mengambil semua property dan semua metode dari kelas lain.

__B.TIPE-TIPE INHERITANCE__




__1.Default__

Default merupakan tipe inheritance dimana property dan method dalam kelas anak (Sub Class) sama dengan kelas induk (Base Class).

Seperti contoh diatas :

    -Class	    = 1. Mahasiswa (Kelas Induk)
                  2. Mahasiswi (Kelas Anak)
    -Method 	= tampilkan (menampilkan nama dan NIM)
    -Property	= Nama, NIM
    -Object 	= Objek (berisikan Nama, NIM)
    Pada contoh diatas kelas mahasiswi mewarisi sifat property & method yang sama dengan kelas mahasiswa. 

__2.Overidding__

Overidding merupakan sebuah tipe inheritance dimana class anak dapat memodifikasi nilai property/method pada class induk. method yang ada dikelas anak harus mempunyai nama yang sama, tanda, dan parameter yang ada dikelas induknya.

Seperti contoh diatas :

    -Class	   = 1.Orang ( Kelas Induk )
                 2. lain ( Kelas Anak )
    -Method    = tampil ( menampilkan nama, profesi, dan alamat )
    -Property  = Nama, profesi, alamat
    -Object    = 1. a= (  berisikan Nama, profesi )
			     2. b= (berisikan Nama, profesi, dan alamat) 
Pada contoh diatas method tampil pada kelas orang memiliki property yang berbeda dengan method tampil pada kelas lain.

__3.Super__

Yang dimaksud dengan super adalah kelas anak dapat mengakses method yang sama dengan kelas induk


Seperti contoh diatas :

    -Class	    = 1. Lingkaran( Kelas Induk )
		          2. kerucut ( Kelas Anak )
    -Method 	= Lingkaran = 1. menampilkan_jari
				              2. luas_permukaan
		          Kerucut   = 1. menampilkan_tinggi
				              2. volume
    -Properti   = Lingkaran = 1. Jari-jari
				              2. Keliling lingkaran
				              3. Luas lingkaran
		          Kerucut   = 1. Tinggi
				              2. volume
    -Object 	= Lingkaran = Obj = menampilkan jari-jari, luas, dan keliling lingkaran
		          Kerucut   = Obj1= menampilkan Jari-jari, tingi, dan volume.

Pada contoh diatas digunakan property super().__init__ untuk mengakses property yang ada di kelas induk.

__4.Multiple__

Multiple inheritance adalah sebuah class yang mengakses dari dua class diatasnya. Seperti contoh jika kita memiliki 2 class A dan B, maka ada class tambahan yakni class C yang mengakses dari class A dan B.

Seperti contoh diatas :

    -Class	  = 1. Murid
		        2. Nilai pelajaran
		        3. Gabungan
    -Method   = Display 
    -Property = 1. Nama
		        2. Math
		        3. Bilology
    -Object   = Objek, Objek2

Pada contoh diatas terdapat kelas gabungan, pada kelas gabungan meng-inherit dari kelas Murid dan NilaiPelajaran untuk menampilkan gabungan antara kelas kelas tersebut.

# METHOD DAN OPERATOR OVERLOADING
__A. PENGERTIAN METHOD OVERLOADING__



Method Overloading adalah kemampuan untuk mendefinisikan suatu method yang sama, dengan nama yang sama tetapi dengan jumlah argumen dan tipe yang berbeda. Termasuk juga kemampuan suatu fungsi untuk melakukan tugas yang berbeda, tergantung pada jumlah parameter atau jenis parameter. Perlu diketahui bahwa tidak semua bahasa pemrograman mendukung method overloading, tetapi Python dapat melakukannya.

Di python jika terdapat dua method yang sama dalam satu kelas maka method yang paling bawah yang menjadi patokan. Untuk itu perlu dilakukan method overloading (Pemanggilan 2 buah method yang sama).

# ENCAPSULATION

__A. PENGERTIAN ENCAPSULATION__


 Encapsulation adalah pembungkus, maksud pembungkus disini adalah untuk menjaga suatu proses program agar tidak dapat diakses secara sembarangan atau di intervensi oleh program lain. Konsep enkapsulasi sangat penting dilakukan untuk menjaga kebutuhan program agar dapat diakses sewaktu-waktu, sekaligus menjaga program tersebut.
 Dalam konsep ini ada dua kata kunci yang digunakan, yakni public, private, dan protected  :
 
- Public memberikan hak akses kepada semua property dan method agar dapat digunakan diluar class tersebut 
- Private memberikan hak akses __hanya__ kepada anggota class tersebut untuk menggunakan dan/atau mengubah nilai dari  property atau method tersebut
- Protected yaitu memberikan hak akses kepada class itu sendiri dan class hasil turunannya (inheritance), artinya apa apa saja yang ada diclass A hanya bisa diakses oleh class A sendiri dan class yang meng Extends class A. Namun harus dipahami class lain yang berada dalam satu package dengan class A mampu mengakses tipe data protected, Sedangkan yang tidak mampu mengakses adalah class-class yang berada diluar package class A. untuk dapat mengaksesnya, class yang berada diluar package class A harus meng extends class A.
